```python
def agefunc(x):
    if x<18:
        return "Child"
    elif x>=18:
        return "Adult"
```

```python
df1["age_group"]=df1["age"].apply(agefunc)
df1
```

```python
def alonefunc(x,y):
    if x or y != 0:
        return False
    else:
        return True
```

```python
df1["is_alone"]=df1.apply(lambda row: alonefunc(row['sibsp'], row['parch']), axis=1)
df1
```
